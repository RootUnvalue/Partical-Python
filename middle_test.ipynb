### interpreator lang vs. compile lang

interpreator lang:
 - 코드 한 줄씩 실행.
 - 목적프로그램 제작 빠름
 - 프로그램 속도 느림

compile lang:
 - 코드를 컴파일러를 통해 한 번에 기계어로 번역.
 - 목적프로그램 제작 느림
 - 프로그램 속도 빠름


```python
print("Hello,World!")
print("Hello," , "World!")
print("\\n \n is new line character")
print("Hello", "World!", sep=", ")
print("Hello", end=", ")
print("World!")

print("can use \\ to escape characters")
print("This is a backslash: \\")
print("This is a double quote: \"")
print("This is a single quote: \'")
print("This is a tab character:\tTabbed")
print("This is a newline character:\nNew line")
print("This is a carriage return character:\rCarriage return")
print("This is a backspace character: a\bBackspace")
print("This is a unicode character: \uAC00")
```

```python
#5
list1 = list()
for i in range(1, 10):
    if i % 2 != 0:
        list1.append(i)
print(list1)

list2 = [i for i in range(1, 10) if i % 2 != 0]
print(list2)

list3 = list()
for i in range(1, 5):
    for j in range(1, 5):
        list3.append(i * j)
print(list3)

list4 = [i + j for i in range(1, 5) for j in range(1, 5)]
print(list4)

list5 = list()
for i, j, k, l in zip(list1, list2, list3, list4):
    list5.append(i + j + k + l)
print(list5)
```

```python
#14
scores = list()
max_s = int(input("학생수: "))
for i in range(0, max_s):
    scores.append(int(input(f"{i+1}번째 학생의 점수 입력: ")))

for i, score in enumerate(scores):
    if score >= 80:
        print(f"{i+1}번 학생은 {score}점으로 합격입니다.")
```

```python
#15
max_score = scores[0]
min_score = scores[0]
for score in scores:
    if score > max_score:
        max_score = score
    if score < min_score:
        min_score = score
print(f"최고 점수: {max_score}")
print(f"최저 점수: {min_score}")
```

```python
#16
scores = [72, 88, 95, 61, 78, 90, 55, 83, 47, 69]

for i, score in enumerate(scores):
    print(f"[{i}] :  {score}")

print()

under60 = 0
for i, score in enumerate(scores):
    if score < 60:
        print(f"[{i}] : {score}")
        under60 += 1
print(f"60점 미만 학생 수: {under60}")

for i, score in enumerate(scores):
    if score >= 90:
        print(f"[{i}] : {score} - 우수")
    elif 90 > score >= 60:
        print(f"[{i}] : {score} - 보통")
    else: 
        print(f"[{i}] : {score} - 미흡")

sum_score = 0
for score in scores:
    sum_score += score
avg_score = sum_score / len(scores)
print(f"평균 점수: {avg_score}")

up80 = [i for i in scores if i >= 80]
for i in up80:
    print(i, end=", ")
```

```python
#18
sales = [45, 52, 38, 61, 73, 89, 94]

i =0
while i < len(sales):
    print(f"{i} : {sales[i]}")
    i += 1

print()

i = 0
while i < len(sales):
    if sales[i] >= 60:
        print(f"{i} : {sales[i]}")
    i += 1

min_sales = sales[0]
max_sales = sales[0]
for sale in sales:
    if sale < min_sales:
        min_sales = sale
    if sale > max_sales:
        max_sales = sale
print(f"최소 매출: {min_sales}")
print(f"최대 매출: {max_sales}")

for i, sale in enumerate(sales, start=1):
    print(f"{i} : {sale}")
```

```python
#19
grades = [[85, 90, 78], [92, 76, 88], [70, 95, 83]]

for i in range(0, len(grades)):
    for j in range(0, len(grades[i])):
        print(f"{grades[i][j]}", end=", ")
    print()

sum_grade_1, sum_grade_2, sum_grade_3 = 0, 0, 0
for i in range(0, len(grades)):
    sum_grade = 0
    for j in range(0, len(grades[i])):
        sum_grade += grades[i][j]
        if i == 0:
            sum_grade_1 += grades[i][j]
        elif i == 1:
            sum_grade_2 += grades[i][j]
        else:
            sum_grade_3 += grades[i][j]
    print(f"{i+1}번째 학생의 평균 점수: {sum_grade / len(grades[i])}")

print(f"국어 평균 점수: {sum_grade_1 / len(grades)}")
print(f"영어 평균 점수: {sum_grade_2 / len(grades)}")
print(f"수학 평균 점수: {sum_grade_3 / len(grades)}")

for i in range(0, len(grades)):
    for j in range(0, len(grades[i])):
        if grades[i][j] >= 90:
            print(f"{i+1}번째 학생의 {j+1}번째 과목 점수: {grades[i][j]}")

ls = [[grade + 5 if grade + 5 <= 100 else 100 for grade in student] for student in grades]
for i in range(0, len(ls)):
    for j in range(0, len(ls[i])):
        print(f"{ls[i][j]}", end=", ")
    print()
```

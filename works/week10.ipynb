```python
#1
a = [10, 20, 30, 40, 50]

i = 0
while i < len(a):
    print(f"{a[i]}", end=", ")
    i += 1

print()

for i in range(len(a)):
    print(f"{a[i]}", end=", ")

print()

for i in a:
    print(f"{i}", end=", ")
```

```python
a = list(range(1, 100))

print("짝수: ", end="")
for i in a:
    if i%2 == 0:
        print(f"{i}", end=", ")

print("\n홀수: ", end="")
for i in a:
    if i%2 == 1:
        print(f"{i}", end=", ")

print("\n3의 배수: ", end="")
for i in a:
    if i%3 == 0:
        print(f"{i}", end=", ")

print("\n10보다 크면서 50보다 작은 수: ", end="")
for i in a:
    if i > 10 and i < 50:
        print(f"{i}", end=", ")

print("\n10보다 작거나 50보다 큰 수: ", end="")
for i in a:
    if i < 10 or i > 50:
        print(f"{i}", end=", ")
```

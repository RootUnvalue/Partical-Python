```python
#10
scores = [88, 92, 79, 93, 85]

#10-1
for i in range(len(scores)):
    if scores[i] >= 80:
        print(scores[i])

#10-2
total = 0
for i in range(len(scores)):
    total += scores[i]
average = total / len(scores)
print(average)

#10-3
max_score = scores[0]
min_score = scores[0]
for i in range(1, len(scores)):
    if scores[i] > max_score:
        max_score = scores[i]
    if scores[i] < min_score:
        min_score = scores[i]
print(max_score)
print(min_score)

#10-4
scores.sort()
print(scores)

#10-5
print(scores[-3:])
```

```python
#11
passengers = []

#11-1
passengers.append([1,1,1,1,1])

#11-2
passengers.append([1,1,1])

#11-3
passengers.append([])
out_3th = passengers[1][2:3].pop()

#11-4
passengers.append([1,1,1,1])

print("현재 승객 리스트", passengers)

s = 0
for i in passengers:
    for j in passengers:
        if j is not []: s = s + 1
print("총 승격 ", s, "명")

print("3번쨰 승강장 데이터: ", out_3th)
print("4번쨰 승강장 데이터: ", passengers[3])
```

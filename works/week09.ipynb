```python
#10
scores = [88, 92, 79, 93, 85]

#10-1
for i in range(len(scores)):
    if scores[i] >= 80:
        print(scores[i], end=', ')

print()

#10-2
total = 0
for i in range(len(scores)):
    total += scores[i]
    
average = 1.0 * total / len(scores)
print("평균: {:.2f}".format(average))

#10-3
max_score = scores[0]
min_score = scores[0]
for i in range(1, len(scores)):
    if scores[i] > max_score:
        max_score = scores[i]
    if scores[i] < min_score:
        min_score = scores[i]
print("최고 점수: {}".format(max_score))
print("최저 점수: {}".format(min_score))

#10-4
scores.sort()
print("정렬 점수 리스트: {}".format(scores))

#10-5
print("상위 3개 점수 리스트: {}".format(scores[-3:]))
```

```python
#21
menu = ('아메리카노', '카페라떼', '카푸치노', '바닐라라떼', '카라멜마키아토')
price = (4500, 5000, 5000, 5500, 6000)

print("1st menu: ", menu[0], "last menu: ", menu[-1])
print("index 1:3 price: ", price[1:4])
print(menu + price)
print((menu * 2))
price = list(price)
for i in range(len(price)):
    price[i] += 500
price = tuple(price)
print(price)
```

```python
#22
student = {'학번': '20251234', '이름': '김파이', '학과': '컴퓨터공학과', '학년': 2, '학점': 3.8}

print("name: ", student['이름'], "\n", "major: ", student['학과'])
# student.update({'전화번호': '010-1234-5678', '이메일': 'kim@example.com'})
student['전화번호'] = '010-1234-5678'
student['이메일'] = 'kim@example.com'
print(student)
print("student key 개수: ", len(student.keys()))

print(student.keys())
print(student.values())
print(student.items())
print(list(zip(student.keys(), student.values())))

scores = dict.fromkeys(['국어', '영어', '수학', '과학'], 0)
print(scores)

for key in student.keys():
    print(key, " : ", student[key])
```

```python
#23
python_club = {'민준', '서연', '도윤', '지우', '하은', '예준'}
java_club = {'서연', '지우', '시우', '수아', '예준', '아인'}

print("합집합" + str(python_club | java_club))
print("교집합" + str(python_club & java_club))
print("차집합(파이썬만 소속))" + str(python_club - java_club))
print("대칭차집합" + str(python_club ^ java_club))
python_club.add("채원")
python_club.add("민준")
print(python_club)
print(len(python_club))

ls = ['사과', '바나나', '사과', '포도', '바나나', '오렌지']
print(ls)
ls = set(ls)
print(ls)
```

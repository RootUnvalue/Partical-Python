```python
#24
def sum_num(a, b):
    '''a, b 두 개의 매게변수를 받고 합을 반환하는 함수'''
    return a + b

def sub_num(a, b):
    '''a, b 두 개의 매게변수를 받고 차를 반환하는 함수'''
    return a - b

def div_num(a, b):
    '''a, b 두 개의 매게변수를 받고 나눗셈 결과를 반환하는 함수'''
    return a / b

def mul_num(a, b):
    '''a, b 두 개의 매게변수를 받고 곱을 반환하는 함수'''
    return a * b

a, b = map(int, input('두 수를 입력하세요: ').split())
print(a, ' + ', b, ' = ', sum_num(a, b))
print(a, ' - ', b, ' = ', sub_num(a, b))
print(a, ' / ', b, ' = ', div_num(a, b))
print(a, ' * ', b, ' = ', mul_num(a, b))

print()
print(sum_num.__doc__)
print(sub_num.__doc__)
print(div_num.__doc__)
print(mul_num.__doc__)
```

```python
#25
def check_word(word):
    return len(word)

def is_all_upper(s):
    if s.upper() == s:
        return True
    else:
        return False

def is_all_lower(s):
    if s.lower() == s:
        return True
    else:
        return False

line = input('문자열을 입력하세요: ')
print('입력한 문자열: ', line)
print('문자열의 길이: ', check_word(line))
print('모두 대문자? : ', is_all_upper(line))
print('모두 소문자? : ', is_all_lower(line))
print('\'python\' 포함? : ', line.find('python') != -1)
print('좌우 공백 제거: ', line.strip())
print('문자열 거꾸로: ', line[::-1])
```

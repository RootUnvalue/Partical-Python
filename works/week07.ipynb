```python
#17
print("x -> y, 썸 or 프로덕트 후 출력")
x = int(input("x: "))
y = int(input("y: "))
act = input("입력 sum 또는 product:")

if act == "sum":
    result = sum(range(x, y + 1))
elif act == "product":
    result = 1
    for i in range(x, y + 1):
        result *= i
print(f"{x} -> {y}, {act}: {result}")
```

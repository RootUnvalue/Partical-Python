```python
# 물건 값 입력
item_price = int(input("물건을 입력하시오: "))

# 동전 개수 입력
coin_1000 = int(input("1000원의 지폐 수를 입력 하시요: "))
coin_500 = int(input("500원의 동전 수를 입력 하시오: "))
coin_100 = int(input("100원의 동전 수를 입력 하시오: "))
coin_10 = int(input("10원의 동전 수를 입력 하시오: "))

# 보유한 총 금액
total_money = coin_1000 * 1000 + coin_500 * 500 + coin_100 * 100 + coin_10 * 10

# 거스름 돈
change = total_money - item_price

print(f"총 거스름 돈은 {change}원입니다. 지급동전은 다음과 같습니다.")

# 거스름 돈을 동전으로 계산
change_1000 = change // 1000
change %= 1000

change_500 = change // 500
change %= 500

change_100 = change // 100
change %= 100

change_10 = change // 10

print(f"지불 거스름 동전은 1000원: {change_1000}개  500원: {change_500}개  100원: {change_100}개  10원: {change_10}개입니다.")
```

```python
import random

# 4개 메뉴의 이름 설정 (기존 변수 a, b, c, d를 메뉴 이름으로 사용)
menu_names = ['a', 'b', 'c', 'd']

# 각 메뉴의 가격을 1000원부터 10000원 사이의 임의의 값으로 설정
menu_prices = {
    'a': random.randint(1000, 10000),
    'b': random.randint(1000, 10000),
    'c': random.randint(1000, 10000),
    'd': random.randint(1000, 10000)
}

print("=== 가게 메뉴 및 가격 ===")
for menu, price in menu_prices.items():
    print(f"{menu}: {price}원")

# 각 메뉴의 판매 개수 입력
print("\n=== 하루 판매량 입력 ===")
sales = {}
for menu in menu_names:
    count = int(input(f"{menu} 메뉴 판매량: "))
    sales[menu] = count

# 총 매출 계산
total_sales = sum(menu_prices[menu] * sales[menu] for menu in menu_names)

# 결과 출력
print("\n=== 판매 현황 ===")
for menu in menu_names:
    menu_total = menu_prices[menu] * sales[menu]
    print(f"{menu}: {sales[menu]}개 × {menu_prices[menu]}원 = {menu_total}원")

print(f"\n=== 하루 총 매출: {total_sales}원 ===")
```
